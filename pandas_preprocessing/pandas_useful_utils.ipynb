 ## nsmallest and nlargest
### This function display 3 smallest and 3 largest values from the dataframe

```python
import pandas as pd
```

```python
df = pd.DataFrame({
...     "Student": ['John', 'Jennifer', 'Jeff', 'Joshua', 'Jack', 'Jeremy', 'Jason'],
...     "Score": [95, 100, 94, 99, 88, 92, 89]
... })
```

```python
df.nlargest(3, "Score")
```

## nsmallest and nlargest
### One time, I needed to find out the largest records by group. For instance, suppose that we have the following DataFrame. I want to extract the records whose scores are the highest in their respective groups.

```python
df = pd.DataFrame({
...     "group": [1, 1, 1, 2, 2, 2, 3, 3, 3],
...     "name": ["A1", "B1", "C1", "A2", "B2", "C2", "A3", "B3", "C3"],
...     "score": [97, 93, 92, 94, 95, 93, 100, 92, 93]
... })
df.head()
```

```python
## Complicated Way : 
df.groupby("group")["score"].max()
```

```python
## Easy Way 
record_indices = df.groupby("group")["score"].idxmax()
record_indices
```

##  qcut
### 
When we deal with continuous variables, it’s sometimes necessary to create ordinal values based on the quantiles of the existing numeric values. <br> Theoretically, some people, including me initially, may tend to find out the quantiles of a certain column and create the categorical column by hard-coding the cutoffs. <br>
However, pandas has an easy option to achieve this operation  <br> — using the qcut function, which create distinct values based on their rankings. Using the DataFrame that we defined in the previous section, <br> we can dichotomize the data based on the Score column, as shown below.


```python
df["score_dichotomized"] = pd.qcut(df["score"], 2, labels=['bottom50', 'top50'])
df
```
